```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
df=pd.read_csv("/kaggle/input/noshowappointments/KaggleV2-May-2016.csv")

# Display first 5 rows to understand the data structure
print(df.head())
```

```python
# Check basic info about the dataset
print("\nDataset Info:")
print(df.info())
```

```python
# Check statistical summary
print("\nStatistical Summary:")
print(df.describe(include='all'))
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Check for duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
#Turn to datetime type necessary columns

datetime_col = ["ScheduledDay","AppointmentDay"]

for col in datetime_col:
    df[col] = pd.to_datetime(df[col]).dt.strftime('%Y-%m-%d')
    df[col] = pd.to_datetime(df[col])

df["ScheduledDay"].head()
```

```python
# Standardize gender column (M/F to Male/Female)
df['Gender'] = df['Gender'].map({'M': 'Male', 'F': 'Female'})

# Standardize no-show column (Yes/No to True/False)
df['No-show'] = df['No-show'].map({'Yes': True, 'No': False})

# Make all column names lowercase and replace hyphens with underscores
df.columns = df.columns.str.lower().str.replace('-', '_')
print("\nStandardized column names:", df.columns.tolist())
```

```python
print("\nCurrent data types:")
print(df.dtypes)
```

```python
# List of columns that should be boolean
bool_cols = ['scholarship', 'hipertension', 'diabetes', 'alcoholism', 'handicap', 'sms_received']

for col in bool_cols:
    if col in df.columns:
        try:
            # Convert to boolean (handling 1/0, 'Yes'/'No', etc.)
            df[col] = df[col].astype(bool)
            print(f"Converted {col} to boolean")
        except Exception as e:
            print(f"Error converting {col} to boolean: {e}")
```

```python
print("\nFinal data types:")
print(df.dtypes)

print("\nSample of converted data:")
print(df.head(3))
```
